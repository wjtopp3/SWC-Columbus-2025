# Introduction to Threat Modeling Concepts
- **Threat modeling** is the process of systematically identifying, evaluating, and addressing potential security threats and vulnerabilities in a system to improve its overall security posture.
- Developing a security mindset means shifting from a builders perspective to an attackers view, seeing software as a collection of components and data flows vulnerable to misuse.
- Threats include not only intentional attacks but also accidents, bugs, hardware failures, and human error.
- Threat modeling helps guide secure decision-making and can be adapted flexibly to fit available time and resources.
- Incremental improvements in threat identification and mitigation significantly strengthen security, even if all vulnerabilities aren't found.
- Threat modeling can also reveal opportunities for non-security-related improvements, such as system efficiencies and new features.
- [1] Kohnfelder Ch. 2

# Identifying Assets and Attack Surfaces
- Identify and prioritize assets based on their value and sensitivity
- Avoid complex risk calculations; instead, use a simple strategy like the Agile "T-shirt size" system (Large, Medium, Small) to prioritize asset protection efforts
  - https://www.easyagile.com/blog/agile-estimation-techniques
- Group similar assets when appropriate for easier management, but separate if  risk profiles or usage contexts differ significantly.
- Always consider asset value from multiple perspectives  including customers, attackers, and the organization itself  to avoid underestimating potential risks.
- Minimize attack surfaces wherever possible, since they are the first points of entry for attackers; early blocking reduces the spread of attacks.
- Recognize that attack surfaces include both digital and physical exposures, such as public network connections, device interfaces, and even side-channel risks like electromagnetic emissions.
